```python
'''
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
'''
```

```python
'''glyphs in bokeh are visual shapes that can be used to represent data on a plot , such as circles ,squre lines retangles'''
```

```python
from bokeh.plotting import figure, show, output_file
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title="Circle Glyphs", x_axis_label="x", y_axis_label="y")
p.circle(x, y, size=20, color="navy", alpha=0.5)


output_file("circle.html")
show(p)
```

```python
'''Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?'''
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
output_file('line.html')
p=figure(title='red plot')
p.scatter(x,y,fill_color='red' ,legend_label='red point')
show(p)
```
